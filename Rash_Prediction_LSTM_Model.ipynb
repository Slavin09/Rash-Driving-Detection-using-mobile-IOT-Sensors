```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical
import joblib

TIME_STEPS = 50
FEATURE_COLS = ["ax","ay","az","azimuth","pitch","roll"]

# Load data
df = pd.read_csv("training_data.csv")
df = df.sort_values("timestamp").reset_index(drop=True)

# Drop the single NaN row
df = df.dropna(subset=["is_rash"])

# Now safe to convert to int
df["label"] = df["is_rash"].astype(int)

# Check class balance
print("Class distribution:")
print(df["label"].value_counts())
print(f"Rash percentage: {df['label'].mean()*100:.2f}%")
print(f"Total samples: {len(df)}")

# Check for NaN in feature columns
print("\nMissing values in features:")
print(df[FEATURE_COLS].isna().sum())

# Drop rows with NaN in features (if any)
df = df.dropna(subset=FEATURE_COLS)

print(f"Samples after cleaning: {len(df)}")

# Scale features
scaler = StandardScaler()
df[FEATURE_COLS] = scaler.fit_transform(df[FEATURE_COLS])

# Create sequences
def create_sequences(data, labels, window):
    X_seq, y_seq = [], []
    for i in range(len(data) - window):
        X_seq.append(data[i:i+window])
        y_seq.append(labels[i+window])
    return np.array(X_seq), np.array(y_seq)

X, y = create_sequences(df[FEATURE_COLS].values, df["label"].values, TIME_STEPS)
y = to_categorical(y, num_classes=2)

print(f"\nSequence shapes: X={X.shape}, y={y.shape}")

# Train-test split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=True, random_state=42,
    stratify=y.argmax(axis=1)  # Stratify by class
)

print(f"Train set: {X_train.shape}, Test set: {X_test.shape}")

# Build model
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(TIME_STEPS, len(FEATURE_COLS))),
    Dropout(0.3),
    LSTM(32),
    Dropout(0.3),
    Dense(2, activation="softmax")
])

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
model.summary()

# Train with early stopping
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True,
    verbose=1
)

history = model.fit(
    X_train, y_train,
    epochs=18,
    batch_size=64,
    validation_split=0.2,
    shuffle=True,
    callbacks=[early_stop]
)

# Evaluate
loss, acc = model.evaluate(X_test, y_test)
print(f"\n{'='*50}")
print(f"✓ Test Loss: {loss:.4f}")
print(f"✓ Test Accuracy: {acc:.4f}")
print(f"{'='*50}")

# Save
model.save("rash_detection_lstm.h5")
joblib.dump(scaler, "imu_scaler.pkl")
print("\n✓ Model saved as 'rash_detection_lstm.h5'")
print("✓ Scaler saved as 'imu_scaler.pkl'")
```

```python
from flask import Flask, request, jsonify
from pyngrok import ngrok
import numpy as np
import joblib
from keras.models import load_model

# -------------------------
# CONFIG
# -------------------------
TIME_STEPS = 50
FEATURE_COLS = ["ax","ay","az","azimuth","pitch","roll"]

# -------------------------
# LOAD MODEL + SCALER
# -------------------------
model = load_model("/content/drive/MyDrive/rash_detection_lstm.h5")
scaler = joblib.load("/content/drive/MyDrive/imu_scaler.pkl")

# sliding window
window = []

# -------------------------
# START NGROK
# -------------------------
ngrok.set_auth_token("2Ta5ArY1tFu9a6wxIzho0HT9fH8_7uhSFzZgpJgNSZMWKS7rM")
public_url = ngrok.connect(5000)
print("NGROK URL:", public_url)

# -------------------------
# FLASK APP
# -------------------------
app = Flask(__name__)

@app.route("/")
def home():
    return "Rash driving predictor running."

@app.route("/predict", methods=["POST"])
def predict():
    global window

    data = request.json
    if not data:
        return jsonify({"error": "empty request"}), 400

    try:
        row = np.array([data[f] for f in FEATURE_COLS], dtype=float)
    except:
        return jsonify({"error": "invalid data fields"}), 400

    # scale
    row = scaler.transform([row])[0]

    # append to window
    window.append(row)

    # if not enough frames → tell client to keep sending
    if len(window) < TIME_STEPS:
        return jsonify({
            "status": "collecting",
            "frames_needed": TIME_STEPS - len(window)
        }), 200

    # prepare sequence
    seq = np.array(window[-TIME_STEPS:])
    seq = seq.reshape(1, TIME_STEPS, len(FEATURE_COLS))

    pred = model.predict(seq, verbose=0)[0]
    rash_prob = float(pred[1])
    rash = int(rash_prob > 0.5)

    return jsonify({
        "status": "ok",
        "rash_probability": rash_prob,
        "rash": rash
    }), 200

# -------------------------
# RUN SERVER
# -------------------------
app.run(port=5000)
```
